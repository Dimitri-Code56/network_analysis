```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an undirected graph
G_undirected = nx.Graph()

# Add nodes
V = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
G_undirected.add_nodes_from(V)

# Add edges
E = [
    ('A', 'B'),
    ('B', 'C'),
    ('B', 'D'),
    ('C', 'D'),
    ('D', 'E'),
    ('D', 'F'),
    ('D', 'G'),
    ('D', 'H'),
    ('G', 'A')
]
G_undirected.add_edges_from(E)

# Add weights to the edges
for edge in G_undirected.edges():
    if edge[::-1] in E:  # Check if the reverse edge exists in the list of edges
        G_undirected[edge[0]][edge[1]]['weight'] = 2
    else:
        G_undirected[edge[0]][edge[1]]['weight'] = 1

# Draw the undirected graph
nx.draw(G_undirected, with_labels=True)

# Display the graph
plt.show()
```
